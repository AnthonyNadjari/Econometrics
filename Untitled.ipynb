```python
import pandas as pd
import numpy as np
data=pd.read_csv("DataLab1.csv", sep=";")
data.head()
```

```python
cac=data["CAC40"]
ml=data["MICHELIN"]
ore=data["LOREAL"]
saf=data["SAFRAN"]
sod=data["SODEXO"]
viv=data["VIVENDI"]
```

```python
data["CAC_ret"]=np.log(cac/cac.shift(1))
data["MICHELIN_ret"]=np.log(ml/ml.shift(1))
data["LOREAL_ret"]=np.log(ore/ore.shift(1))
data["SAFRAN_ret"]=np.log(saf/saf.shift(1))
data["SODEXO_ret"]=np.log(sod/sod.shift(1))
data["VIVENDI_ret"]=np.log(viv/viv.shift(1))
```

```python
est_return_cac = np.mean(data["CAC_ret"])
est_return_ml = np.mean(data["MICHELIN_ret"])
est_return_ore = np.mean(data["LOREAL_ret"])
est_return_saf = np.mean(data["SAFRAN_ret"])
est_return_viv = np.mean(data["VIVENDI_ret"])

est_return_cac_ann=est_return_cac *252 
est_return_ml_ann=est_return_ml *252 
est_return_ore_ann=est_return_ore *252 
est_return_saf_ann=est_return_saf *252 
est_return_viv_ann=est_return_viv *252 

est_vol_cac = np.std(data["CAC_ret"])
est_vol_ml = np.std(data["MICHELIN_ret"])
est_vol_ore = np.std(data["LOREAL_ret"])
est_vol_saf = np.std(data["SAFRAN_ret"])
est_vol_viv = np.std(data["VIVENDI_ret"])

est_vol_cac_ann = np.std(data["CAC_ret"])*np.sqrt(252)
est_vol_ml_ann = np.std(data["MICHELIN_ret"])*np.sqrt(252)
est_vol_ore_ann = np.std(data["LOREAL_ret"])*np.sqrt(252)
est_vol_saf_ann = np.std(data["SAFRAN_ret"])*np.sqrt(252)
est_vol_viv_ann = np.std(data["VIVENDI_ret"])*np.sqrt(252)
    
print(est_return_cac)
```
